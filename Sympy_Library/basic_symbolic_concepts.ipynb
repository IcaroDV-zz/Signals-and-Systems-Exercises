```python
import sympy as sp 
```

```python
x = sp.symbols('x') #   TOPIC 1
                    #
expre1 = x**2       #   if you want to work with symbolic variables,
expre1              #   you must first declare them as such.
```

```python
a = 3               #   if you don't do that, the result is going to be
expre2 = a**2       #   numeric
expre2
```

```python
                                           #  TOPIC 2               
                                           #  
y, z = sp.symbols('y z')                   #  if you want to explicit 
expre3 = sp.Eq(y,(3*x**2 + 3*x)/3 - 4/z)   #  equations in implict 
expre3                                     #  form, so you must do it                                            #  like this
```

```python
#   TOPIC 3          
#                                                  
#   if your intention is to plot some 2 dimension graphics, you can do #   it this way
sp.plot(expre1)
#   but you may notice that if you put expre3 instead of expre1 in 
#   sp.plot(), it may not work both because of expre3 is in implicit 
#   form and because it has 3 dimensions.
```

```python
expre1 = sp.Eq(y,x**2)      #   
sp.plot_implicit(expre1)    #   a) solving implicit form problem
```

```python
from sympy.plotting import plot3d   #  b) one way to solve 2          
expre4 = (x**2 + 0.07*y**3)         #  dimensions problem
plot3d(expre4)
```

```python
#   TOPIC 4
#
#   for solving simple equations you have
sp.solve(expre4)
```

```python
#   Solving a linear system
eq1 = sp.Eq(3 * x + 4 * y, 4)
eq2 = sp.Eq(4 * x + 3 * y, 3)

eqs = [eq1,eq2]

sp.linsolve(eqs,(x,y))
```

```python
#   TOPIC 5
#
#   Replacing symbols non permanently
expre1.subs(x,z**2)
```

```python
#   Replacing symbols permanently
expre = expre1.subs(x, z**2)
expre
```

```python
expr =  sp.sin(x)**2 + sp.cos(x)**2
expr
```

```python
#   Simplifying eqs
sp.simplify(expr)
```

```python
#   Expanding eqs
expre5 = z*(x + y)
sp.expand(expre5)
```

```python
#   Factoring eqs
expre6 = x * y + z * y
sp.factor(expre6)
```
